```python
def calculate_magnitude_distance(inputTable, metricCol, storeComparison):
    """
    Calculate a standardised magnitude distance for a metric
    between a trial store and each control store.

    Parameters:
    - inputTable: DataFrame containing columns [STORE_NBR, YEARMONTH, metricCol]
    - metricCol: str, name of the column with the metric
    - storeComparison: int/str, the trial store ID

    Returns:
    - DataFrame with mean magnitude similarity per store pair
      Columns: [Store1, Store2, mag_measure]
    """
    storeNumbers = inputTable['STORE_NBR'].unique()
    trial_data = inputTable[inputTable['STORE_NBR'] == storeComparison][['YEARMONTH', metricCol]]

    # Collect differences between trial and each control store
    calcDistTable = []
    for store in storeNumbers:
        control_data = inputTable[inputTable['STORE_NBR'] == store][['YEARMONTH', metricCol]]

        # Merge trial & control data by YEARMONTH
        merged = pd.merge(trial_data, control_data, on="YEARMONTH", suffixes=("_trial", "_control"))
        merged['Store1'] = storeComparison
        merged['Store2'] = store
        merged['measure'] = (merged[f"{metricCol}_trial"] - merged[f"{metricCol}_control"]).abs()

        calcDistTable.append(merged[['Store1', 'Store2', 'YEARMONTH', 'measure']])

    calcDistTable = pd.concat(calcDistTable, ignore_index=True)

    # Standardise distances per (Store1, YEARMONTH)
    minMaxDist = (
        calcDistTable.groupby(['Store1', 'YEARMONTH'])['measure']
        .agg(minDist='min', maxDist='max')
        .reset_index()
    )
    distTable = pd.merge(calcDistTable, minMaxDist, on=['Store1', 'YEARMONTH'])
    distTable['magnitudeMeasure'] = 1 - (distTable['measure'] - distTable['minDist']) / (
        distTable['maxDist'] - distTable['minDist']
    )

    # Final average per store pair
    finalDistTable = (
        distTable.groupby(['Store1', 'Store2'])['magnitudeMeasure']
        .mean()
        .reset_index()
        .rename(columns={'magnitudeMeasure': 'mag_measure'})
    )

    return finalDistTable
```

```python
trial_store = 77

# --- Correlations ---
corr_TOTSales = calculate_correlation(preTrialMeasures, "totSales", trial_store)
corr_nCustomers = calculate_correlation(preTrialMeasures, "nCustomers", trial_store)

# --- Magnitude distances ---
magnitude_nSales = calculate_magnitude_distance(preTrialMeasures, "totSales", trial_store)
magnitude_nCustomers = calculate_magnitude_distance(preTrialMeasures, "nCustomers", trial_store)

# --- Combine correlation + magnitude into composite scores ---

# Merge for sales
score_nSales = (
    pd.merge(corr_TOTSales, magnitude_nSales, on=["Store1", "Store2"])
    .assign(scoreNSales=lambda x: 0.5 * x["corr_measure"] + 0.5 * x["mag_measure"])
)

# Merge for customers
score_nCustomers = (
    pd.merge(corr_nCustomers, magnitude_nCustomers, on=["Store1", "Store2"])
    .assign(scoreNCust=lambda x: 0.5 * x["corr_measure"] + 0.5 * x["mag_measure"])
)

# --- Composite score across sales + customers ---
composite_score = (
    pd.merge(score_nSales[['Store1','Store2','scoreNSales']],
             score_nCustomers[['Store1','Store2','scoreNCust']],
             on=['Store1','Store2'])
    .assign(compositeScore=lambda x: 0.5 * x["scoreNSales"] + 0.5 * x["scoreNCust"])
    .sort_values("compositeScore", ascending=False)
)

# Show top candidate control stores
print(composite_score.head())
```
