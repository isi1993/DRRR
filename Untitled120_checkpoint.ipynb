```python
#from scipy.stats import ttest_ind

#  Perform a t-test to see if the difference is significant.
data_copy = merged

# Filter only midage and young singles/couples
subset = data_copy[data_copy['LIFESTAGE'].isin(['MIDAGE SINGLES/COUPLES', 'YOUNG SINGLES/COUPLES'])]

# Function to run independent t-test between two groups
def run_ttest(data, lifestage, group1, group2):
    # Subset for a given lifestage
    stage_data = data[data['LIFESTAGE'] == lifestage]

    # Extract sales/units (replace 'TOT_SALES' with the right metric column)
    g1 = stage_data[stage_data['PREMIUM_CUSTOMER'] == group1]['TOT_SALES']
    g2 = stage_data[stage_data['PREMIUM_CUSTOMER'] == group2]['TOT_SALES']

    # Perform t-test
    t_stat, p_val = ttest_ind(g1, g2, equal_var=False, nan_policy='omit')
    print(f"Lifestage: {lifestage} | {group1} vs {group2} -> t-stat: {t_stat:.4f}, p-value: {p_val:.4f}")

# Run t-tests for both lifestages
for stage in ['MIDAGE SINGLES/COUPLES', 'YOUNG SINGLES/COUPLES']:
    run_ttest(subset, stage, 'Mainstream', 'Premium')
    run_ttest(subset, stage, 'Mainstream', 'Budget')
    run_ttest(subset, stage, 'Premium', 'Budget')
```
